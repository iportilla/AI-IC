# Prime Number Calculation

This notebook shows two ways to calculate prime numbers up to a limit `n`:

1. Long division (trial division)
2. Sieve of Eratosthenes

Both versions include comments so the logic is easy to follow.

## Method 1: Long Division

For each number, we test whether it can be divided evenly by any smaller candidate factor.

This is straightforward to understand, but it becomes slower as the limit grows.

```python
def is_prime_long_division(number):
    # Numbers smaller than 2 are not prime.
    if number < 2:
        return False

    # We only need to check divisors up to the square root of the number.
    limit = int(number ** 0.5) + 1

    for divisor in range(2, limit):
        # If the remainder is 0, the number has a factor and is not prime.
        if number % divisor == 0:
            return False

    # No divisor was found, so the number is prime.
    return True


def primes_by_long_division(n):
    # Build a list of all prime numbers from 2 through n.
    return [value for value in range(2, n + 1) if is_prime_long_division(value)]
```

## Method 2: Sieve of Eratosthenes

The sieve marks multiples of each prime number as composite.

This is usually much faster than checking each number with long division when `n` is large.

```python
def primes_by_sieve(n):
    # There are no prime numbers below 2.
    if n < 2:
        return []

    # Start by assuming every number is prime.
    is_prime = [True] * (n + 1)
    is_prime[0] = False
    is_prime[1] = False

    # Only iterate up to the square root because smaller primes
    # will already have removed the needed multiples.
    limit = int(n ** 0.5) + 1
    for number in range(2, limit):
        if is_prime[number]:
            # Mark all multiples of the current prime as not prime.
            for multiple in range(number * number, n + 1, number):
                is_prime[multiple] = False

    # Collect the numbers that are still marked as prime.
    return [value for value in range(2, n + 1) if is_prime[value]]
```

## Example Run

This cell calculates prime numbers up to a chosen limit with both methods and checks that the results match.

```python
limit = 50

long_division_primes = primes_by_long_division(limit)
sieve_primes = primes_by_sieve(limit)

print(f"Primes up to {limit} using long division: {long_division_primes}")
print(f"Primes up to {limit} using sieve: {sieve_primes}")
print(f"Results match: {long_division_primes == sieve_primes}")
```

## Performance Example

This example uses a much larger limit of 1,000,000 to show the runtime difference between long division and the Sieve of Eratosthenes.

To keep the output readable, it prints the number of primes found and the elapsed time for each method.

```python
from time import perf_counter

performance_limit = 1_000_000

# Measure long division performance on a large input.
start_time = perf_counter()
long_division_primes = primes_by_long_division(performance_limit)
long_division_time = perf_counter() - start_time

# Measure sieve performance on the same input.
start_time = perf_counter()
sieve_primes = primes_by_sieve(performance_limit)
sieve_time = perf_counter() - start_time

print(f"Prime count up to {performance_limit}: {len(sieve_primes)}")
print(f"Long division time: {long_division_time:.4f} seconds")
print(f"Sieve time: {sieve_time:.4f} seconds")
print(f"Results match: {long_division_primes == sieve_primes}")
```

## Efficiency Plot

This chart compares runtime as the limit grows, so it is easier to see how quickly the sieve pulls away from long division.

The final benchmark point uses a limit of 1,000,000.

```python
import matplotlib.pyplot as plt
from time import perf_counter

benchmark_limits = [1_000, 10_000, 100_000, 1_000_000]
long_division_times = []
sieve_times = []

for benchmark_limit in benchmark_limits:
    # Time the long division approach for the current limit.
    start_time = perf_counter()
    long_division_result = primes_by_long_division(benchmark_limit)
    long_division_times.append(perf_counter() - start_time)

    # Time the sieve approach for the same limit.
    start_time = perf_counter()
    sieve_result = primes_by_sieve(benchmark_limit)
    sieve_times.append(perf_counter() - start_time)

    # Confirm both methods still produce the same primes.
    print(f"Limit {benchmark_limit}: results match = {long_division_result == sieve_result}")

plt.figure(figsize=(10, 6))
plt.plot(benchmark_limits, long_division_times, marker='o', linewidth=2, label='Long division')
plt.plot(benchmark_limits, sieve_times, marker='o', linewidth=2, label='Sieve of Eratosthenes')
plt.xscale('log')
plt.yscale('log')
plt.title('Prime Generation Performance Comparison')
plt.xlabel('Limit')
plt.ylabel('Execution time (seconds)')
plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.legend()
plt.show()
```
