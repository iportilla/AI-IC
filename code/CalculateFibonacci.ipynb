# Fibonacci Examples
This notebook defines a simple Fibonacci function and runs a few example inputs.

```python
def fibonacci(n):
    """Return the nth Fibonacci number. Negative input returns 0."""
    if n < 0:
        return 0
    if n <= 1:
        return n

    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b
```

```python
examples = [0, 1, 2, 3, 4, 5, 10, -1]
for value in examples:
    print(f"fibonacci({value}) = {fibonacci(value)}")
```
